```python
rank["country_full"] = rank["country_full"].str.replace("IR Iran", "Iran").str.replace("Korea Republic", "South Korea").str.replace("USA", "United States")
```

```python
rank = rank.set_index(['rank_date']).groupby(['country_full'], group_keys=False).resample('D').first().fillna(method='ffill').reset_index()
```

```python
temp = pre_game.merge(rank[["country_full", "total_points", "previous_points", "rank", "rank_change", "rank_date"]], left_on=["date", "home_team"], right_on=["rank_date", "country_full"]).drop(["rank_date", "country_full"], axis=1)

rank = temp.merge(rank[["country_full", "total_points", "previous_points", "rank", "rank_change", "rank_date"]], left_on=["date", "away_team"], right_on=["rank_date", "country_full"], suffixes=("_home", "_away")).drop(["rank_date", "country_full"], axis=1)
```

```python
rank[(rank.home_team == "Brazil") | (rank.away_team == "Brazil")].tail(10)
```

```python
def result_finder(home, away):
    if home > away:
        return pd.Series([0, 3, 0])
    if home < away:
        return pd.Series([1, 0, 3])
    else:
        return pd.Series([2, 1, 1])

results = rank.apply(lambda x: result_finder(x["home_score"], x["away_score"]), axis=1)
```

```python
rank[["result", "home_team_points", "away_team_points"]] = results
```

```python
rank
```

```python
rank["rank_dif"] = rank["rank_home"] - rank["rank_away"]
rank["score_dif"] = rank["home_score"] - rank["away_score"]
rank["points_home_by_rank"] = rank["home_team_points"]/rank["rank_away"]
rank["points_away_by_rank"] = rank["away_team_points"]/rank["rank_home"]
rank["relative_point_dif"] = rank["points_home_by_rank"] - rank["points_away_by_rank"]
rank["score_dif_by_rank_dif"] = rank["score_dif"]/(rank["rank_home"]-rank["rank_away"])
```

```python
stats_val = []

for index, row in team_stats.iterrows():
    team = row["team"]
    date = row["date"]
    past_games = team_stats.loc[(team_stats["team"] == team) & (team_stats["date"] < date)].sort_values(by=['date'], ascending=False)
    last10 = past_games.head(10)
    last5 = past_games.head(5)
    last3 = past_games.head(3)
    
    goals = past_games["score"].mean()
    goals_10 = last10["score"].mean()
    goals_5 = last5["score"].mean()
    goals_3 = last3["score"].mean()
    
    goals_suf = past_games["suf_score"].mean()
    goals_suf_10 = last10["suf_score"].mean()
    goals_suf_5 = last5["suf_score"].mean()
    goals_suf_3 = last3["suf_score"].mean()
    
    ranks = past_games["rank_suf"].mean()
    ranks_10 = last10["rank_suf"].mean()
    ranks_5 = last5["rank_suf"].mean()
    ranks_3 = last3["rank_suf"].mean()
    
    if len(last10) > 0:
        points = past_games["total_points"].values[0] - past_games["total_points"].values[-1]
        points_10 = last10["total_points"].values[0] - last10["total_points"].values[-1] 
        points_5 = last5["total_points"].values[0] - last5["total_points"].values[-1] 
        points_3 = last3["total_points"].values[0] - last3["total_points"].values[-1] 
    else:
        points = 0
        points_10 = 0
        points_5 = 0
        points_3 = 0
        
    tpoints = past_games["team_points"].mean()
    tpoints_10 = last10["team_points"].mean()
    tpoints_5 = last5["team_points"].mean()
    tpoints_3 = last3["team_points"].mean()
    
    points_rank = past_games["points_by_rank"].mean()
    points_rank_10 = last10["points_by_rank"].mean()
    points_rank_5 = last5["points_by_rank"].mean()
    points_rank_3 = last3["points_by_rank"].mean()
    
    score_rank = past_games["score_dif_by_rank_dif"].mean()
    score_rank_10 = last10["score_dif_by_rank_dif"].mean()
    score_rank_5 = last5["score_dif_by_rank_dif"].mean()
    score_rank_3 = last3["score_dif_by_rank_dif"].mean()

    stats_val.append([goals, goals_10,goals_5,goals_3, goals_suf, goals_suf_10, goals_suf_5,goals_suf_3,
                      ranks, ranks_10, ranks_5,ranks_3,points, points_10,points_5, points_3,  
                      tpoints, tpoints_10, tpoints_5,tpoints_3,
                      points_rank, points_rank_10, points_rank_5,points_rank_3,
                      score_rank, score_rank_10,score_rank_5, score_rank_3 ])
```
